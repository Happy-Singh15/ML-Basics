```python
import pandas as pd
import numpy as np
```

```python
time = pd.read_csv('../../../data/messages.csv')
```

```python
time.head()
```

```python
time.drop(columns='msg',inplace=True)
```

```python
time.head()
```

```python
time.info()
```

```python
# converting to datetime datatype

time['date'] = pd.to_datetime(time['date'])
```

```python
time.info()
```

```python
time['hour'] = time['date'].dt.hour
time['minute'] = time['date'].dt.minute
time['second'] = time['date'].dt.second

time.sample(5)
```

```python
# Extract only time part form datetime

time['time'] = time['date'].dt.time

time.head()
```

```python
# time difference

import datetime

today = datetime.datetime.today()

today
```

```python
today-time['date']
```

```python
# time differnce in hours

time['hour_diff']= (today-time['date'])/np.timedelta64(1,'h')

time.head()
```
