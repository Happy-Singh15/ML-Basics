```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/train.csv')
```

```python
df.head()
```

# 1. Categorical Data

a. Countplot

```python
sns.countplot(x ='Survived',data =df,palette='viridis')
```

```python
sns.countplot(x='Pclass',data=df,palette='viridis')
```

```python
sns.countplot(x='Sex',data=df,palette='viridis')
```

```python
order = df['Embarked'].value_counts().index
sns.countplot(x='Embarked',data=df,hue='Embarked',palette='viridis',order=order,legend=True)
```

b. Pie Chart

```python
df['Survived'].value_counts().plot(kind='pie',x='Survived',autopct='%1.1f%%')
```

```python
df['Pclass'].value_counts().plot(kind='pie',x='Pclass',autopct='%.2f%%')
```

# 2. Numerical Data

a. Histogram

```python
# plt.hist(df['Age'],bins=10)
sns.histplot(df['Age'],bins=10)
```

b. Distplot

```python
sns.displot(df['Age'],kind='kde')
```

c. Boxplot

```python
sns.boxplot(df['Age'],orient='h')
```

```python
sns.boxplot(df['Fare'],orient='h')
```
